```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm # for scientific computing
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
#The kind of plot to draw
sns.catplot(x='day',y='total_bill',kind='box',data=tips,hue='sex')
```

```python
# use to set style of background of plot
sns.set(style="whitegrid")
sns.boxplot(x ='tip', y ='day', data = tips)
```

```python
order = ['Sun', 'Thur', 'Fri', 'Sat']

sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, order = order)
```

```python
hue_order = ['Female', 'Male']


sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, hue_order = hue_order)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, orient='v')
```

```python
sns.boxplot(x = 'total_bill', y = 'size', hue = 'sex',
           data = tips, orient='h')
```

```python
# set color
sns.boxplot(x = 'day', y = 'tip', data = tips, color="g")
          
```

```python
#palette is used to set the color of the plot
# set color map or palette
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, palette="rocket_r") 
```

```python
sns.boxplot(x ='sex', y ='total_bill', data = tips, palette ='plasma')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, saturation=0.12)
```

```python
### dodge= When hue nesting is used, whether elements should be shifted along the categorical axis.
                                         
# shift hue categorical variable bar bar in one bar
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, dodge= False)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips, dodge= True) 
```

```python
## Draw outlines around the data points using linewidth:
# Width of the gray lines that frame the plot elements. Whenever we increase linewidth than the
# point also will increase automatically.
sns.boxplot(x = 'day', y = 'tip',  data = tips, linewidth=2.5)
                      
               
```
