## Lineplot

A single line plot presents data on x-y axis using a line joining datapoints. To obtain a graph Seaborn comes with an inbuilt function to draw a line plot called lineplot().

Syntax: lineplot(x,y,data)

where,

xâ€“ data variable for x-axis

y- data variable for y-axis

data- data to be plotted

```python
url = 'https://raw.githubusercontent.com/sanketchavan5595/EDA-on-bestselling-books-on-Amazon-/main/bestsellers%20with%20categories.csv'
bestsellers_df = pd.read_csv(url)
bestsellers_df.head()
```

```python
bestsellers_df.info()
```

```python
bestsellers_df.isnull().sum()
```

```python
bestsellers_df.shape #(There are total of 550 books listed as bestsellers on amazon)
```

Categorical variables
- Genre
- Year

Numerical variables:
  - Reviews
  - User Rating
  - Price 

```python
# hue=used for categorical data that  will produce line different color

plt.figure(figsize=(7,7))

sns.lineplot(x = 'Price', y = 'Reviews', hue = 'Genre',data = bestsellers_df);

plt.title("Price v/s Reviews")
```

Fiction book have more reviews than the non-fiction book but non-fiction books have higher prices

```python
plt.figure(figsize = (16,9))
sns.set(style='darkgrid',)

# Draw line plot of size and total_bill with parameters and hue "day"
sns.lineplot(x = "Year", y = "User Rating", data = bestsellers_df, hue = "Genre",
            style = "Genre", palette = "hot", dashes = False,size=0.10,
            markers = ["o", "*"],  legend="brief",)  ## o-circle,*,p,P=star,pentagon,Plus

plt.title("Line Plot", fontsize = 20)
plt.xlabel("Size", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.show()
```

Attributes: 

    Context: plotting context parameters
    Style: defines style
    Palette: sets a color palette
    Font
    Font_scale: sets font size
    Color_codes: If set True then palette is activated, short hand notations for colors can be remapped  from the palette.

## Histogram

# Scatter plot

scatterplot() function is used to plot the data and depict the relationship between the values using the scatter visualization.

seaborn.scatterplot(*, x=None, y=None, hue=None, style=None, size=None, data=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, x_bins=None, y_bins=None, units=None, estimator=None, ci=95, n_boot=1000, alpha=None, x_jitter=None, y_jitter=None, legend='auto', ax=None, **kwargs)

From the above observation we can see that its a comparision between who and fare based on the condition weather the passenger is alive and he is a man, woman or child.

my observation says that woman survived the most of all then comes the child and the last man survived

```python
sns.scatterplot(x = 'age', y ='fare',hue='who', data=df,style='who',alpha=1, legend='brief')
```

From this observation we can see that age group less than 15 years are child and from the age group between 20 to soon all other than child come.

## cat plot

catplot shows frequencies (or optionally fractions or percents) of the categories of one, two or three categorical variables. 

seaborn.catplot(*, x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, seed=None, order=None, hue_order=None, row_order=None, col_order=None, kind='strip', height=5, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# dodge= When hue nesting is used, whether elements should be shifted along the categorical axis.
sns.catplot(x='day',y='total_bill',data=df,hue='sex',kind='box',dodge=True, estimator=np.mean)
```

From the above observation we can see that male people pay more bills as compare to women an all the week days. we are are also constructing the 5 number summary based on mean.

```python
sns.catplot(x='day',y='total_bill',data=df,hue='sex',kind='box',dodge=False,estimator=np.median)
```

From the above observation we can see that male people pay more bills as compare to women an all the week days. we are are also constructing the 5 number summary based on median.
