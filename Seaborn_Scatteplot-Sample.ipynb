# How to create Sebaorn Scatter Plot?

```python
# Import libraries
import seaborn as sns # for Data visualization
import matplotlib.pyplot as plt # for Data visualization
import pandas as pd
```

```python
sns.get_dataset_names()
```

 It contains information of all the passengers aboard the Titanic, which unfortunately was any embark_town. 
    This dataset can be used to predict whether a given passenger survived or not. 

```python
#Import dataset 
titanic_df=sns.load_dataset("titanic")
titanic_df
```

```python
titanic_df.columns
```

```python
##Checking the NULL values 
titanic_df.isnull().sum()
```

```python
# passengerid:- Uniqueid of the passenger
#survived:-If the passenger survived(0-No, 1-yes)
# Pclass: 
# 1st = Upper
# 2nd = Middle
# 3rd = Lower
# Name- Name of the passenger
#sex- Male/female
#Age: passenger age in year
# Sibsp: The dataset defines family relations in this way...
# Sibling = brother, sister, stepbrother, stepsister
# Spouse = husband, wife (mistresses and fiancÃ©s were ignored)
# Parch: The dataset defines family relations in this way.............................
# Parent = mother, father
# Child = daughter, son, stepdaughter, stepson
# Some children travelled only with a nanny, therefore parch=0 for them.
#ticket=ticket number
#fare:passenger fare
# cabin: cabin number
#embark: to board a ship 
# alive=It depends on what kind of survival analysis
```

```python
# import dataset from folder

#DataFrame_name = pd.read_csv("file name of file path")
```

```python
sns.scatterplot(x ="sex", y="age", data = titanic_df)
```

```python
sns.scatterplot(x="sex",y="age",hue ="survived",data=titanic_df)
```

```python
# Draw Seaborn Scatter Plot to find relationship between age and fare
sns.scatterplot(x="age",y="fare", data=titanic_df)

# or -- you can pass DataFrame x and y variable directly
#sns.scatterplot(x = titanic_df.age, y = titanic_df.fare)
```

```python
sns.scatterplot(x = titanic_df.age, y = titanic_df.fare)
```

```python
sns.scatterplot(x=titanic_df['age'],y=titanic_df["fare"])
```

# hue

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex")
```

# hue_order

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",hue_order=["female","male"])
```

# size & sizes

```python
#Grouping variable that will produce points with different sizes.Can be either categorical or numeric
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who")
```

```python
#The minimum and maximum size to use such that other values are normalized within this range.
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who",sizes=(50,300))
```

# size_order

```python
#size_order : list Specified order for appearance of the ``size`` variable levels, otherwise they are determined from the data.
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who",size_order=['child','man','woman'],)
```

# size_norm

# style 

```python
#style : vector or key in ``data``Grouping variable that will produce points with different markers.
sns.scatterplot(x='age',y='fare',data=titanic_df,style="who",)  ##  style="column_name"
```

# style_order

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,style="who",style_order=['child','woman','man'])
```

# palette

```python
#Colors to use for the different levels of the hue variable.
sns.scatterplot(x='age',y='fare',data=titanic_df,hue="sex",palette="hot")                                                          
                # palette does not work without hue
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,palette="hot") 
```

# alpha

```python
#alpha : float Proportional opacity of the points.
sns.scatterplot(x='age',y='fare',data=titanic_df,alpha=0.7)
```

# Legend

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',legend=False)
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',legend=True)
```

# ax - axes

```python
ax=sns.scatterplot(x='sex',y='age',data=titanic_df)
ax.set(xlabel="Sex",
      ylabel="Age",
      title="Seaborn scatter plot of Sex and Age")
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='sex',y='age',data=titanic_df,hue="alive",style="alive",size="who",sizes=(100,400))
plt.show()
```
